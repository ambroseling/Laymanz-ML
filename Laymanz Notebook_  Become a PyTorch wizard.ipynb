```python
import tensorflow as tf
import torch
```

```python
import time
```

```python
SIZE = 1024
a_tf = tf.random.uniform(shape=[SIZE,SIZE])
b_tf = tf.random.uniform(shape=[SIZE,SIZE])
start_tf = time.monotonic()
c_tf = tf.matmul(a_tf,b_tf)
end_tf = time.monotonic()
```

```python

a_th = torch.randn(SIZE,SIZE).to(torch.device("cuda"))
b_th = torch.randn(SIZE,SIZE).to(torch.device("cuda"))
start_th = time.monotonic()
c_th = a_th @ b_th
end_th = time.monotonic()
```
