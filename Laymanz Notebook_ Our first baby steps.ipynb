# Probability & Statistics üéØ

1. Probability Basics
Probability Theory: The study of randomness and uncertainty. Understanding probability helps in modeling the likelihood of events happening.
Conditional Probability: How the probability of an event changes when another related event is known to occur. This is crucial in many ML algorithms, like Naive Bayes.
Bayes' Theorem: A way to reverse conditional probabilities that's essential in many ML contexts, such as spam filtering and disease diagnosis.


2. Statistical Functions
Mean and Median: Measures of central tendency that summarize the central point of a data distribution.
Variance and Standard Deviation: Measures of dispersion that describe the spread of data points around the mean.
Covariance and Correlation: Indicate the relationship and dependency between two variables. Correlation coefficients are used to quantify the strength of a relationship.

3. Distributions
Normal Distribution (Gaussian): A very common continuous distribution used in statistics because many techniques and methods are optimized to work with it.
Binomial Distribution: Useful for modeling the number of successes in a fixed number of independent Bernoulli trials (like flipping a coin).
Poisson Distribution: Often used for modeling the number of times an event happens in a fixed interval of time or space.

4. Hypothesis Testing
Null and Alternative Hypotheses: Formulating and testing assumptions about a population using sample data.
P-values and Significance Levels: Used to determine the strength of the evidence against the null hypothesis.
Confidence Intervals: Provide a range of values that likely contain the population parameter.

## Bayesian Statistics
Bayesian statistics offers a powerful framework for machine learning (ML) by incorporating **prior knowledge, handling uncertainty explicitly, and offering probabilistic predictions.**


Condintional probability is at the very heart of Bayesian statistics. It works by leveraging a set of observations denoted as $X$, lets assume that the observations are generated by a distribution $P$.

When we say that observations are generated by a distribution $P$, it means that the data points in our "observation" are considered to be **random samples drawn from that specific probability distribution.** That probability distribution could be normal, biased, etc. Depends on the scenario.

Examples:
- If each data point in my observation represents the height of an individual adult man. Then the observations (heights) are assumed to follow a normal distribution.
- if each data point represents the outcome of a single die roll.
Then since the die is fair, each face (1 through 6) has an equal probability of landing face up. This situation is described by a uniform distribution.

We denote the parameters of this distribution as $H$, and lets also assume we know the shape of this distribution.

**Bayes rule:**
$$
P(H|X) = \frac{P(X|H) \cdot P(H)}{P(X)}
$$
where


* $P(H|X)$: posterior distribution (based on an observation, whats the probability of )
* $P(X|H)$: likelihood (based on some set of parameters, whats the probability of observation)
* $P(H):$ prior distribution (distribution of parameters)
* $P(X):$ marginalization or evidence (the probability of the observations being true)

**NOTE**:
We assume:
* posterior has a certain shape
* parameters are drawn from a given prior

Ali Intuition:
Posterior distribution: having a prior beleif about the scenario. Example: you are trying to see if a **coin is biased or not**, one way is by doing a lot of coin flips. Obviously, the more coin flips you do, the more certain you will be if the coin is biased or not.

My question is: **How do we model this change in certainty, as we do more coin flips?**

I start off by assuming the coin is not biased, so my posterior distrutibution, at 0 coin flips will be a uniform distribution, with equal likelihood of a heads or a tails. This is my Prior Belief at this point in time (0 flips).
As I observe a heads, or a tails on a coin flip, I continue to "update" this beleif. Lets see what this "updated" prior Belief might look like:

Notice that in this scenario, the coin ended up being a fair coin. (at 500 tosses, 250 of them are heads). If it was actually biased, lets say Probability of a heads is .7, then my Prior will STILL be uniform, and would eventually get updated to being .7 for Probability of heads. Something like this:


image.png

The key idea here is how our prior distribution is able to model our observations at each coin flip. So it incorporates Prior Knowledge, and handles uncertainty (if we have done 1 flip and seen 1 heads, it doesn't mean the probabilty of tails is 0).

Lets model this:


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import Button, VBox, Output, IntText, FloatSlider
from scipy.stats import beta
import time

# Set up the initial parameters for the Beta distribution
alpha_param, beta_param = 1, 1  # Uniform prior

# Create output widget to display the plot and text output for flip results
out_plot = Output()

# Create input field, button, and reset button
n_flips_text = IntText(value=10, description='Number of Flips:', layout={'width': '180px'})
prob_slider = FloatSlider(value=0.5, min=0, max=1, step=0.01, description='Prob Heads:', readout_format='.2f')
button_auto_flip = Button(description='Start Auto Flip', button_style='primary', layout={'width': '150px', 'height': '50px'})
button_reset = Button(description='Reset', button_style='danger', layout={'width': '100px', 'height': '50px'})

# Function to update the plot based on the coin flip
def update_plot(heads, tails):
    x = np.linspace(0, 1, 100)
    y = beta.pdf(x, 1 + heads, 1 + tails) # This is where the
    with out_plot:
        out_plot.clear_output(wait=True)
        plt.figure(figsize=(10, 5))
        plt.plot(x, y, label=f'Heads: {heads}, Tails: {tails}')
        plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
        plt.title("Bayesian updating of posterior probabilities")
        plt.xlabel("$p$, probability of heads")
        plt.ylabel("Density")
        plt.legend()
        plt.show()

# Handlers for automatic flipping and reset
heads_count = 0
tails_count = 0

def on_auto_flip_clicked(b):
    global heads_count, tails_count
    n_flips = n_flips_text.value
    for _ in range(n_flips):
        if np.random.rand() < prob_slider.value:
            heads_count += 1
        else:
            tails_count += 1
        update_plot(heads_count, tails_count)
        time.sleep(0.5)  # Pause for half a second between flips for visibility

def on_reset_clicked(b):
    global heads_count, tails_count
    heads_count = 0
    tails_count = 0
    update_plot(heads_count, tails_count)

# Attach the button click events to their respective handlers
button_auto_flip.on_click(on_auto_flip_clicked)
button_reset.on_click(on_reset_clicked)

# Display the widgets and output
display(VBox([n_flips_text, prob_slider, button_auto_flip, button_reset]), out_plot)

# Initial plot
update_plot(heads_count, tails_count)
```
