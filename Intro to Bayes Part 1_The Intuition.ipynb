## Brief history of Bayesian Statistics

Bayesian statistics is named after Thomas Bayes, an 18th-century mathematician, who developed the initial ideas of [probability inversion.](https://chatgpt.com/share/6fe67326-0a0b-4ef6-8f25-f6d60a21e62b)

Bayes' Theorem, published after the death of Thomas in 1763, provided the foundation for Bayesian inference. So he never actually got to see his work, and the true potential it had.


The 20th century worked towards the more formal Bayesian methods, with significant contributions from statisticians like Pierre-Simon Laplace, who generalized Bayes' ideas and applied them to various scientific fields.


## What is Bayesian Statistics?


Bayesian Statistics expresses a degree of belief in an event. This belief may change as new evidence is presented. The Bayesian approach uses Bayes' Theorem to update the probability of a hypothesis as more data or information becomes available.

In short, Bayesian statistics combines prior knowledge with new evidence to form updated, more accurate predictions with the present data.

### The difference between Bayes and Conditional Probability:


Conditional Probability refers to the probability of an event occurring given that another event has already occurred.
The formula is simply:
$$P(A|B) = \frac{P(A \text{ and } B)}{P(B)}$$

While Bayes theorum is a way of finding a probability when other related probabilities are known (like likelihood, Posterior etc).
And the formula, is a tweaked version of Conditional Probability:
$$P(A|B) = \frac{P(A) *P(B|A)}{P(B)}$$

## 1: Explaining what a Prior is



In simple words, a prior is your initial belief or estimate about a "parameter" **before you see new data**.

It represents what you think you know about the parameter based on previous obervations/data, knowledge, experience, and/or assumptions. The **parameter** in most cases is simply the probability of an event, given some other new observed data.

Biased coin example:

Imagine you have a coin, and you want to know the probability that it lands heads (let's call this probability $Î¸$).

**Before Flipping the Coin:**

Suppose you have no reason to believe the coin is biased, so you might initially believe that heads and tails are equally likely.
This belief can be represented as a prior distribution where $Î¸$ is equally likely to be any value between 0 and 1, but you might center your belief around 0.5 (since you believe the coin is fair).

**After Flipping the Coin:**

You start flipping the coin and observe the outcomes. Based on these outcomes, you update your belief about $Î¸$. This updated belief, which combines your prior with the new data, is called the posterior distribution.

## 3: Explaining what Likelihood is.

Likelihood is not the same thing as probability. It does not have to add up to 1.

I like to think of Likelihood, as us trying to best Estimate, what the UNDERLYING distribution, from which the data came from, looks like.


We use it, when we OBSERVE the data, and want to know, whats the probability that the DISTRIBUTION the data came from, was with a certain parameter. Lets discuss this futher:

Lets say you flip this coin twice, without knowing if it is biased or not.

if you see that it landed 2 times on heads,
- whats the probability of this happening, if the **coin was fair**? $P(2\text{ heads} |\text{ probability of heads} = .5)$ = it is: $.5*.5$ = $.25$. Therefore, the likelihood of obsreving 2 heads, if the coin is fair is $.25$
- whats the probability of this happening, if the coin was NOT fair, and was in reality biased towards heads by .7?:

 Well then the Likelihood of this happening would be: $P(2\text{ heads} |\text{ probability of heads} = .7)$ = it is: $.7*.7$ = $.49$.

This "probability" is actually the likelihood, of observing specific results, under the assumption that the coin is either biased (with a certain $Î¸$ value, or not.

$P(\text{Data/results we wish to see }|\text{ given a certain hypothesis (AKA a fixed value for Î¸)})$

## 4: Explaining what "Normalizing Constant" means



For this, we can come back to Conditional Probability, and why we have to normalize in the first place.

When we are calculating probability, GIVEN some other fact, we have to account for that right.

For example:

Example: Suppose you have 100 students:

- 20 of them study more than 10 hours a week.
- 5 of those who study more than 10 hours get an A.

Therefore, P(A and B) = 5/100 = 0.05 (5%).
P(B) = 20/100 = 0.20 (20%).

If I ask you now, calculate how many students will get an A, OUT of the 20 students who study more then 10 hours.
Thinking about it logically, we know that out of the 20 students, 5 get an A, so the probability must be: $\frac{5}{20}$
This evaluates to the expression:

$$P(\text{get an A | Study more then 10 hours}) = \frac{P(A \text{ and } B)}{P(B)}$$
Which gives us $\frac{0.05}{0.20}$ = .25.
Hence, by normalizing,
- we are accounting for the fact that we have reduced our Sample space of probabilities we are looking at.
- And we ensure the probabilities add up to 1.
