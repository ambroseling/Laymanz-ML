# Chapter 1: Overview of ML systems

ML systems learn complex patterns from existing data and use these patterns to make predictions on unseen data

When to use ML:
1) When the system has the capacity to learn, when there is data for it to learn (i.e. predicting rental prices of AirBnB places)
2) When there are complex patterns to learn (stock prices, crypto prices, object detection, speech recognition)
    - Traditional software: Inputs, Pattern -> output
    - ML: Inputs, Outputs -> pattern
3) There is data and you can collect it
    - you need ML algos to learn from data
    - zero shot learning: use an ML system  
    - continual learning: deploy a model and learn from incomign data in production
4) Its a prediction problem
    - make approximations with ML
5) Unseen data share patterns with training data
    - temporal relevancy of the data is very important (data from 10 years ago vs rn)
    - you can make assumtions about user behaviour
6) Its repetitive 
    - if it is repetitive there should be a pattern to be learnt from it
7) Cost of wrong predictions is cheap
    - wrong predictions wont have catastrophic consequences
8) Can be scaled
    - can make a lot of predictions (inferences) at the same time
    

**ML in research**
- SOTA performance
- fast training high throughput
- data is static
- dont care about fairness & interpretability


**ML in production**
- different stakeholders have different requirements
- fast inference and low latency
- data is constantly changing
- data in production is very unstructed, messy, noisy, imbalanced
- must consider fairness and interpretability 

**NOTE**:
* latency: how long it takes to receive 1 query
* throughput: how many queries can you spit out in a time frame
* batched queries: can mean higher latency AND throughout
* latency matterns a lot in production

# Data Engineering

### Data sources
- user input data
- system generated data (system logs)


### Data format
- JSON: text
- CSV: text
- Porquet: binary
- Avro: binary primary
- Protobuf: binary primary

#### JSON
- langauge indepedent
- key value pair paradigm
- json files are text files
- take up a lot of space

#### Row Major VS Column Major
Column Major 
- data is stored and retrieved column by column
- good for accessing **features**
- better when you need to do a lot of column-based reads

Row Major: 
- data is stored and retrieve row by row
- good fo raccessing **samples**
- allow faster data writes


#### Text VS binary
- binary files are more compact
- text files are interpretable but larger


### Data formats

- Relational Model (SQL)
    - Data is organized into **relations**
    - relations are a **set of tuples**
    - In a relation the order of columns or rows dont matter
    - we usually **normalize** relations
    - in SQL you specify the data pattern you want
    - query optimizers figure out how to execute the query
- No SQL
    - document model
        - JSON,XML, BSON
        - no enforced schema, schemaless
        - shifts the responsibility of assuming structrues from the application that writes the data to the application that reads the data
        - better locality than relational model, easier to retrieve certain info
        - but executing joins is tough compared to relational models (joins are when )
    - graph model
        - relatinoship between data items are the priority


### Structured VS Unstructured data
- structured data 
    - has predefined schema
    - easy to search and analyze
    - can only handle data with a specific schema
    - stored in data warehouses
- unstructured data 
    - is flexible wrt storage options, how you read it
    - fast arrival
    - handle data from any source
    - no need to worry about schema changes
    
- **Data Warehouse**: storing structured data
- **Data lake**: storing unstructured data


#### Data processing
- Transactional and Analytical Processing
    - Online transcational Processing
        - need to be fast, no latency
        - if your system cant process the transcation, it shouldnt go through
        - **ACID**:
            - Atomicity: all steps in the transcation are completed successfully as a group
            - If any step in the transaction fails, all other steps fail too
        - Consistency:
            - all transcations coming throguh must follow predefined rules
        - Isolation:
            - 2 transcations happening simultaneously have to happen indepdently
        - Durability:
            - gurantee that a transcation has been **commited** it will not change and it will stay commited 
        - Transcational DB are row major
    - Online Analytical Processing
        - storage and processing are tightly coupled
        - ETL (Extract Transform and Load)
        - Extract: we extract from different sources
            - Data can be malformed, corrueted, needs validation of data
        - Tranform:
            - data processing
            - join data from mulitple sources
        - Load:
            - hwo to load your transformed data into target destination

#### Data flow
- through db, services (REST APIs, RPC APIs), real time transport (Apache, Kafka)
